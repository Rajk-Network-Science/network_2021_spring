# 4.Házi: Skálafüggetlen hálozatok

Az első körben definiálok egy gráf objetumot. A gráf egy ER gráf, ahol 20 csúcs élei 20 százalékos valószínűséggel kapcsolódnak.

```python
import networkx as nx
import hazi_4 as hazi
G=nx.erdos_renyi_graph(20,0.2,seed=10)
```

A Barabári-Albert modell egy olyan növekedésű modell, melyhez az új csúcsok a létező csúcsokhoz, a létező csúcsok fokszámával lineárisan arányosan kapcsolódnak

```python
hazi.BA_model(G)
```

Itt látható a gráf fokszámeloszlása. Látható, hogy pár node az összes kapcsolat jelentős számát birtoloja, majd sok node rendelkezik alacsony fokszámmal.

```python
hazi.k_distrib(G, fit_line=True, expct_lo=1, expct_hi=100, expct_const=10)
```

Így néz ki a gráf, amelyben szintén látható, hogy 20-30 igazán nagy fokszámmal rendelkező node van (a csúcsok méretei a fokszámmal arányosan növekednek)

```python
hazi.vizual(G)
```
