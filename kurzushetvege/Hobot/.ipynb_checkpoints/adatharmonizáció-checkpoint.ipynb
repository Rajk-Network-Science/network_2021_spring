## 2010-es Adat beolvasása

```python
df2010=pd.read_excel("bilat2010.xlsx", skiprows=1)
```

```python
df2010=df2010.set_index(df2010.columns[0]).stack().reset_index()
df2010.columns=["Origin", "Destination", "2010"]
```

### országnevek harmonizálása

```python
[cn for cn in df.Origin.unique() if cn not in df2010.Origin.unique()]
```

```python
[cn for cn in df2010.Origin.unique() if cn not in df.Origin.unique()]
```

```python
rename_dict={'Hong Kong, China':'Hong Kong SAR, China'}
```

```python
for col in ["Origin", "Destination"]:
    df2010[col]=df2010[col].replace(rename_dict)
```

```python
df2010
```

```python
df=df.merge(df2010, how="left", left_on=["Origin", "Destination"], right_on=["Origin", "Destination"])
```

```python
def calc_flow(group):
    
    group["Quantity_1"] = group["Quantity"].shift(1).fillna(0)
    group["Flow"] =group["Quantity"].astype(int) - group["Quantity_1"].astype(int)
    return group

#flow_df = pd.concat([calc_flow(group[1]) for group in mina.groupby(["Origin", "Destination"])])
#flow_df
```
