```python
import networkx as nx
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import math
```

### [Stanley Milgram - The small-world problem](http://snap.stanford.edu/class/cs224w-readings/milgram67smallworld.pdf)

##### Is this figure below an accurate representation of social networks? Why (not)?

```python
def circle(r, center, n_points):
    return [
        [
            math.cos(2 * math.pi / n_points * x) * r + center[0],
            math.sin(2 * math.pi / n_points * x) * r + center[1],
        ]
        for x in range(n_points)
    ]
```

```python
def branching_circles(levels, center, size, n):

    circle_dict = {"0_0": [np.array(center)]}

    for l in range(levels):

        origin = np.concatenate(
            [circle_dict[k] for k in circle_dict.keys() if k[0] == str(l)]
        )

        new_points = np.concatenate([circle(size / (l + 1) ** 2, c, n) for c in origin])

        for i, p in enumerate(new_points):

            circle_dict[f"{l + 1}_{i}"] = [p]

    return circle_dict
```

```python
levels = 2
center = [0, 0]
size = 1
n = 10
```

```python
branch = {k: v[0] for k, v in branching_circles(levels, center, size, n).items()}
```

```python
colors = {
    k: "red" if k[0] == "0" else "blue" if k[0] == "1" else "orange"
    for k in branch.keys()
}
```

```python
g = nx.Graph()
g.add_nodes_from(branch.keys())
```

```python
for key in branch.keys():

    k = int(key[0])

    if k:
        for i in range(n ** (k - 1)):

            g.add_edges_from(
                [[f"{k - 1}_{i}", f"{k}_{j}"] for j in range(i * n, (i + 1) * n)]
            )
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
ax.axis("off")
nx.draw_networkx_nodes(
    g,
    branch,
    node_size=200,
    node_color="white",
    edgecolors=list(colors.values()),
    linewidths=5,
)
nx.draw_networkx_edges(g, branch, width=2, edge_color="grey")
plt.show()
```
